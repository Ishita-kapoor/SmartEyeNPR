```python
import cv2
import os
import matplotlib.pyplot as plt

folder_custom = "/content/drive/MyDrive/dataset_npr/images_npr_yolo/custom_dataset/images"
folder_kaggle = "/content/drive/MyDrive/dataset_npr/images_npr_yolo/kaggle_dataset/cv_project_data/images"

def resized(path):
    print(f"Processing folder: {path}")
    for filename in os.listdir(path):
        print(f"Found file: {filename}")

        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            print("Skipped (not an image)")
            continue

        path_img = os.path.join(path, filename)
        img = cv2.imread(path_img)

        if img is None:
            print(f"‚ùå Failed to read image: {filename}")
            continue

        print(f"‚úÖ Loaded: {filename} | Shape: {img.shape}")

        if img.shape[0:2] != (640, 640):
            res = cv2.resize(img, (640, 640))
            cv2.imwrite(path_img, res)  # Save resized image
            print(f"üìè Resized: {filename}")
resized(folder_custom)
resized(folder_kaggle)
```
